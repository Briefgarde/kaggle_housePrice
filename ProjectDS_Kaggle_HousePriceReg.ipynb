### Procèdure actuelle 

La procédure d'évaluation immobilière d'AgileEstate comprend plusieurs étapes, débutant par une prise de contact avec le vendeur potentiel et se poursuivant par une visite du bien immobilier. Ensuite, une analyse des données est effectuée au sein de l'agence, en s'appuyant sur les informations historiques issues de l'entrepôt de données. Un retour détaillé est fourni au vendeur, généralement le jour même ou le lendemain. Cette procédure vise à déterminer une fourchette de valeurs dans laquelle un bien peut être vendu, en se référant à l'offre et à la demande du secteur, ainsi qu'aux attributs spécifiques du bien tels que sa situation géographique. *La procédure actuelle est détaillée ci-dessous :*

**1. Prise de contact initiale**

- Réceptionner la demande de vente d'un bien immobilier, par téléphone ou par email.

**2. Visite sur place**

- L'agent immobilier se rend sur place pour procéder à une évaluation détaillée du bien.
- *Inspection minutieuse de la propriété, incluant* :

    - Documenter les caractéristiques spécifiques (matériaux, installations).

    - Discuter avec le propriétaire pour obtenir des informations supplémentaires.
    - Relever les données géographiques.

**3. Retour à l'agence**
- Saisir manuelle des informations dans le logiciel de gestion de biens immobiliers.
- Crééer un fichier Excel à partir de l'application pour pouvoir réaliser l'évaluation du prix du bien.

**4. Consultation de l'environnement de vente**
- Examiner les biens similaires actuellement en vente, comparables et situés dans la même zone géographique.
- Consulter les données issuent de l'entrepôt de données de l'agence AgileEstate, lequel contient des informations historiques détaillées sur les transactions immobilières.

**5. Définition de la fourchette de prix**
- Définir une fourchette de prix comprenant une estimation basse, moyenne et haute.
- Justifier cette fourchette en indiquant les critères utilisés.

**6. Finalisation**
- Fournir un rapport d'estimation détaillé au propriétaire.

# Creation du modèle baseline

## Creation de la colonne utilisable pour le résultat voulu.

Afin de mesurer les performances de nos modèle, il est nécessaire de définir un métrique. Dans nos cas, nous avons choisi d'utiliser le Root Mean Squared Error (RMSE) entre le logarithm de prix de vente réel provenant du dataset et le logarithme du prix de vente estimé par notre modèle. 

Les raisons sont les suivantes :

1) **Adaptabilité du RMSE :** Le RMSE est une mesure particulièrement adaptée pour les tâches de régression car il donne une indication claire de l'erreur moyenne en termes d'unités dans le jeu de données. Cette métrique est sensible aux erreurs importantes car elle élève les erreurs au carré avant de prendre la racine carrée, ce qui la rend très utile pour identifier les modèles qui pourraient sous-performer de manière significative sur certains échantillons du jeu de données. Le RMSE peut donc aider à optimiser notre modèle en mettant l'accent sur la minimisation des grandes erreurs, ce qui est crucial dans des applications pratiques comme l'estimation des prix immobiliers où ces grandes erreurs peuvent être très coûteuses.

2) **Utilité de la prise du logarithme :** L'utilisation du logarithme des prix de vente dans le calcul du RMSE a plusieurs avantages significatifs. D'abord, elle aide à normaliser la distribution des prix, qui est souvent asymétrique et étalée, en une distribution plus symétrique et serrée, réduisant ainsi l'influence des valeurs extrêmes ou des outliers. Cela rend l'évaluation plus robuste et moins sensible aux variations extrêmes de prix, qui pourraient autrement dominer l'erreur moyenne calculée. De plus, en comparant les logarithmes, nous mesurons les erreurs en termes de pourcentages plutôt qu'en termes de valeurs absolues, ce qui est plus pertinent quand les gammes de prix varient largement entre les observations, ce qui est souvent le cas dans l'immobilier. Cela signifie que le RMSE sur le log des prix nous donne une idée plus précise de l'erreur relative, permettant une évaluation plus équitable entre des biens de différentes valeurs.

```python
train['LogSalePrice'] = np.log(train['SalePrice'])

# si nécessaire, vous pouvez obtenir le prix de vente réel a partir du log de cette manière : 
# Ce code permet de reverser le logarithme, et est donc utilisé lors de l'envoi des résultats sur Kaggle. 
# train['SalePrice'] = np.exp(train['LogSalePrice'])
```

```python
# Current main baseline model

numerical_cols = ['YearBuilt', 'LotArea', 'GrLivArea', 'TotRmsAbvGrd']
categorical_cols = ['BldgType', 'Neighborhood']

# Step 2: Preprocessing
# For numerical features: scale
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# For categorical features: apply one-hot encoding
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessors in a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Step 3: Create the regression pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Step 4: Splitting the data into training and testing sets
all_cols = numerical_cols + categorical_cols
X = train[all_cols]
y = train['LogSalePrice']
# Cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)

# Print the results
print("Cross-validated RMSE scores:", rmse_scores)
print("Mean RMSE:", np.mean(rmse_scores))
```

#### Feature engineering : 

Le feature engineering est un processus essentiel en data science qui consiste à créer de nouvelles colonnes (ou features) à partir des colonnes existantes dans un dataset. Cette étape est cruciale car elle permet d'extraire des informations pertinentes et utiles qui ne sont pas immédiatement évidentes à partir des données brutes. En générant des nouvelles features, on peut capturer des relations complexes et des patterns qui améliorent la capacité des modèles de machine learning à faire des prédictions précises. Ces nouvelles features peuvent enrichir les données et offrir de meilleures perspectives aux modèles. Le feature engineering permet donc d'augmenter la performance des modèles, de réduire les erreurs et d'apporter des insights plus profonds et plus précis, rendant le modèle plus efficace et pertinent.

```python
def house_age_at_sale(df):
    df['AgeAtSale'] = df['YrSold'] - df['YearBuilt']
    return df

def time_since_remod(df):
    df['YearsSinceRemodel'] = df['YrSold'] - df['YearRemodAdd']
    return df
```

```python
feature_engineering_transformers = [
    ('house_age_at_sale', FunctionTransformer(house_age_at_sale)),
    ('time_since_remod', FunctionTransformer(time_since_remod)),
]

# Create a custom transformer to apply the feature engineering transformations
class FeatureEngineering(BaseEstimator, TransformerMixin):
    def __init__(self, transformers):
        self.transformers = transformers

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        for name, transformer in self.transformers:
            X = transformer.transform(X)
        return X

# Instantiate the custom transformer
feature_engineering_step = FeatureEngineering(feature_engineering_transformers)
```

## Pipeline

Notre pipeline est faite en plusieurs étapes. 

1. En premier, nous faisons les étapes de features engineering. Nous créeons des features supplémentaires qui peuvent apporter des informations supplémentaires pour le modèle. 
2. Ensuite, nous définissons comment sont traité les données manquantes. Sur la base du dictionnaire défini plus haut, chacune des catégories (numérique, ordinal ou catégorique) est traité d'une de deux manière : Soit une valeur en dure en mise à la place de la valeur manquante, soit la valeur manquante est remplacé par une moyenne des autres data points. 
    1. Pour les valeurs numériques, la valeur en dur est 0. Dans le dataset, nous avons vu que les valeurs nulles répresentes souvent que la feature n'est pas la. Par exemple, GarageCar est parfois nulles. Cela veut dire que la maison n'a pas de garage, et donc que le nombre de voiture qu'il peut contenir est 0. 
    2. Pour les valeurs catégorical, ordinale ou non, la valeur en dur est 'NA' (non-ordinal) ou 'Missing' (ordinal). Comme pour les valeurs numériques, lorsque ces valeurs sont utilisé, cela veut dire que la maison ne possède pas les éléments nécessaire pour entrer dans une des catégories défini dans le dataset. Le fait que deux valeurs ('NA' et 'Missing') sont utilisés provient simplement du fait que notre système n'a pas fonctionné lorsque 'NA' a été utilisé. 'Missing' était nécessaire pour fonctionner. 
3. Une fois que la dataframe ne contient plus aucun valeur nulles, nous pouvons effectuer le OneHotEncode sur les colonnes catégoriques (ordinale ou non). Dans le cas des valeurs ordinales, nous utilisons plutôt un OrdinalEncoder, car il permet de préserver la notion d'ordre dans les catégories.  
4. Finalement, nous utilisons un StandardScaler pour applanir le tous et obtenir de meilleures valeurs pour le modèle. Un StandardScaler transforme la dataframe pour que toutes les valeurs d'une colonne soient centrées sur 0, donc la valeur d'une ligne pour la colonne en question correspond a l'écart entre 0 et soi-même. La plupart des modèles performe mieux lorsqu'un Scaler est utilisé.

### Entrainer un modèle RandomForestRegressor

```python

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Définition du scorer RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# Fonction objective pour Optuna
def objective(trial):
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 5, 10),
        'max_depth': trial.suggest_int('max_depth', 3, 30),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2']),
        'bootstrap': bootstrap,
    }
    
    # Ajouter max_samples seulement si bootstrap est True
    if bootstrap:
        param['max_samples'] = trial.suggest_float('max_samples', 0.5, 1.0)
    
    # Création et entraînement du modèle
    model = RandomForestRegressor(
        random_state=42,
        **param
    )
    
    # Utilisation de cross_val_score pour la validation croisée
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Retourne la moyenne du RMSE
    return np.mean(scores)

# Création d'une étude et optimisation de la fonction objective
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=200)  # Augmenter le nombre d'essais

# Affichage des meilleurs paramètres
print('Best trial:')
trial = study.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```

# Entrainer un modele Support Vector Regression (SVR)

```python
import optuna
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.svm import SVR
import numpy as np

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Définition du scorer RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# Fonction objective pour Optuna
def objective(trial):
    param = {
        'C': trial.suggest_loguniform('C', 1e-3, 1e3),
        'epsilon': trial.suggest_loguniform('epsilon', 1e-3, 1e1),
        'kernel': trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid']),
    }
    
    # Si le kernel est poly ou rbf, nous devons ajouter le paramètre gamma
    if param['kernel'] in ['poly', 'rbf', 'sigmoid']:
        param['gamma'] = trial.suggest_categorical('gamma', ['scale', 'auto'])
    
    # Si le kernel est poly, nous devons également ajouter le paramètre degree
    if param['kernel'] == 'poly':
        param['degree'] = trial.suggest_int('degree', 2, 5)
    
    # Création et entraînement du modèle
    model = SVR(**param)
    
    # Utilisation de cross_val_score pour la validation croisée
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Retourne la moyenne du RMSE
    return np.mean(scores)

# Création d'une étude et optimisation de la fonction objective
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)

# Affichage des meilleurs paramètres
print('Best trial:')
trial = study.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```
