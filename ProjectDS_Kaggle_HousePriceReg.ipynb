## Contexte

AgileEstate, une entreprise familiale suisse de renom spécialisée dans le secteur immobilier, offre divers services, y compris un service d'estimation de la valeur des biens. Malgré un parcours riche et une réputation bien établie, l'entreprise fait face à une concurrence accrue et accuse un certain retard technologique par rapport à des agences plus innovantes.

***Importance des estimations immobilières***

- **Attractivité pour les clients** : L'estimation gratuite attire un large éventail de clients.
- **Précision de l'estimation** : Une estimation précise permet de vendre rapidement et au juste prix, évitant ainsi de "brader" le bien ou de prolonger le processus de vente.
- **Personnalisation nécessaire** : Chaque bien est unique et nécessite une estimation personnalisée pour garantir une évaluation juste, adaptée aux spécificités de la propriété.


***Concurrence digitalisée***

Neho, une des agences immobilières les mieux digitalisées, se distingue par une approche transparente et équitable. Ils offrent une solution numérique complète et moins coûteuse, avec une évaluation en ligne suivie par l'intervention d'experts à certaines étapes clés du processus.

***Acteurs clés actuels***

- **Agents immobiliers :** Professionnels responsables de la vente et de l'achat de biens immobiliers.
- **Responsables de l'agence :** Dirigeants assurant la gestion stratégique et opérationnelle.
- **Clients vendeurs :** Propriétaires souhaitant vendre leurs propriétés.
- **Clients acheteurs :** Acquéreurs potentiels recherchant des biens immobiliers.
- **Équipe de data analysts :** Spécialistes de la gestion et de l'analyse des données.

### Procédure d’estimation actuelle 

La procédure d'évaluation immobilière d'AgileEstate comprend plusieurs étapes, débutant par une prise de contact avec le vendeur potentiel et se poursuivant par une visite du bien immobilier. Ensuite, une analyse des données est effectuée au sein de l'agence, en s'appuyant sur les informations historiques issues de l'entrepôt de données. Un retour détaillé est fourni au vendeur, généralement le jour même où le lendemain. Cette procédure vise à déterminer une fourchette de valeurs dans laquelle un bien peut être vendu, en se référant à l'offre et à la demande du secteur, ainsi qu'aux attributs spécifiques du bien tels que sa situation géographique. *La procédure actuelle est détaillée ci-dessous :*

**Prise de contact initiale**
- Réceptionner la demande de vente d'un bien immobilier, par téléphone ou par email.
- Planifier un rendez-vous avec le potentiel vendeur afin de visiter le bien et de l'évaluer.

**Visite sur place**
- L'agent immobilier se rend sur place pour procéder à une évaluation détaillée du bien.
- Inspection minutieuse de la propriété, incluant :
- Prise de mesures précises.
- Prise de notes manuscrites sur l'état général du bien.
- Documenter les caractéristiques spécifiques (matériaux, installations).
- Photographier les pièces et les éléments clés.
- Discuter avec le propriétaire pour obtenir des informations supplémentaires.
- Relever les données géographiques.

**Retour à l'agence**
- Saisir manuelle des informations dans le logiciel de gestion de biens immobiliers.
- Créer un fichier Excel à partir de l'application pour pouvoir réaliser l'évaluation du prix du bien.

**Consultation de l'environnement de vente**
- Examiner les biens similaires actuellement en vente, comparables et situés dans la même zone géographique.
- Consulter les données issues de l'entrepôt de données de l'agence AgileEstate, lequel contient des informations historiques détaillées sur les transactions immobilières.

**Définition de la fourchette de prix**
- Définir une fourchette de prix comprenant une estimation basse, moyenne et haute.
- Justifier cette fourchette en indiquant les critères utilisés.

**Finalisation**
- Fournir un rapport d'estimation détaillé au propriétaire.


La procédure actuelle est complexe est chronophage. Le service d'estimation de AgileEstate peine à se moderniser, et a subi une forte baisse de demande, des clients qui préfères utiliser d'autres services que ceux offert ici. L'émergence d'acteurs digitalisés comme Neho a bouleversé le marché immobilier. Les estimations précises et rapides sont désormais indispensables pour les entreprises désireuses de rester efficaces et de réduire les coûts. Ces technologies permettent des estimations plus rapides et performantes, offrant ainsi aux agences un gain de temps considérable et une réduction des coûts de commission. Si AgileEstate ne s'adapte pas rapidement, des agences digitales comme Neho risquent de supplanter sa position sur le marché.

## Problématique
Récemment, les dirigeants d'AgileEstate ont constaté une baisse significative des demandes pour leurs services d'estimation, au profit de services moins chers et plus rapides. Une analyse approfondie a révélé les raisons suivantes :

**1. Concurrence accrue des sites d'estimation en ligne** : Les clients obtiennent des estimations suffisamment fiables via des sites tels que Zillow ou Neho. Ils ne voient plus l'intérêt d'obtenir une estimation professionnelle qui peut parfois être lente et fastidieuse.

**2. Longueur du processus d'estimation** : Actuellement, le processus prend plus d'une journée, un délai jugé trop long par les clients.

**3. Expérience client insatisfaisante** : Certains clients ne sont pas satisfaits des estimations, qui prennent du temps et offrent une fourchette de prix trop large par rapport à la concurrence qui utilise des technologies de pointe.

**4. Processus de vente pas assez rapide** : Les estimations n'étant pas toujours fiables, les biens ont tendance à être surévalués ou sous-évalués, créant une latence ou une perte au niveau des ventes.

**5. Non-exploitation des données** : Bien que l'entrepôt de données permette d'obtenir des insights intéressants, il n'est pas suffisamment exploité pour améliorer les prédictions de prix.

### Distribution des prix de vente (colonne cible)

Nous commençons par analyser la distribution des prix de vente des maisons, ce qui nous donnera une première idée des tendances et des anomalies possibles dans les données.

```python
plt.figure(figsize=(18, 6))

# First subplot for 'SalePrice'
plt.subplot(1, 2, 1)
sns.histplot(train['SalePrice'], bins=100)
plt.xlabel('Prix de vente ($)', fontsize=14)
plt.ylabel('Frequence', fontsize=14)
plt.title('Distribution des prix de ventes', fontdict={'fontsize': 11, 'fontweight': 'bold'})

# Second subplot for 'LogSalePrice'
plt.subplot(1, 2, 2)
logSalePrice = np.log(train['SalePrice'])
sns.histplot(logSalePrice, bins=100)
plt.xlabel('Logarithme du prix de vente', fontsize=14)
plt.ylabel('Frequence', fontsize=14)
plt.title('Distribution du log des prix de vente', fontdict={'fontsize': 11, 'fontweight': 'bold'})

plt.tight_layout()
plt.show()
```

Le diagramme de gauche montre la distribution des prix des maisons. On observe une concentration élevée de maisons dont les prix se situent entre 100'000 et 300'000 francs. Les maisons dont les prix dépassent cette fourchette sont beaucoup moins nombreuses, avec un déclin progressif jusqu'à environ 750'000 francs.

Le diagramme de droite représente la distribution des prix transformés en logarithmes. Cette transformation permet de réduire la variabilité et de normaliser la distribution des prix, la rendant plus symétrique et proche d'une distribution normale. Cette transformation est utile pour les analyses statistiques car elle atténue l'effet des prix extrêmes et permet une meilleure modélisation des données.

### Relation entre les variables numériques et le prix de vente

Analyser la corrélation entre les variables numériques et le prix de vente nous aide à identifier les caractéristiques ayant le plus d'impact sur la valeur des maisons.

```python
# Calculer la matrice de corrélation pour l'ensemble de données d'entraînement en prenant en compte uniquement les colonnes numériques
corr = train.corr(numeric_only=True)

# Trier les valeurs de corrélation par rapport à 'SalePrice' dans l'ordre décroissant
corr_sale = corr['SalePrice'].sort_values(ascending=False)

# Définir la taille de la figure pour le graphique
plt.figure(figsize=(18, 14))

# Créer un graphique à barres horizontales des corrélations entre les variables et le prix de vente
plt.barh(corr_sale.index, corr_sale.values)

# Ajouter un label à l'axe des x
plt.xlabel("Correlation", size=12)

# Laisser l'axe des y sans label
plt.ylabel("")

# Ajouter un titre au graphique
plt.title("Relation entre les variables numériques et le prix de vente", fontdict={'fontsize': 11, 'fontweight': 'bold'})

# Inverser l'ordre des éléments sur l'axe des y pour que les variables avec la plus haute corrélation soient en haut
plt.gca().invert_yaxis()

# Afficher le graphique
plt.show()
```

### Points clés

Les points clés à noter dans ce graphique sont :

- `OverallQual` (Qualité globale) est la variable ayant la plus forte corrélation positive avec le prix de vente.
- `GrLivArea` (Surface habitable au-dessus du niveau du sol) et `GarageCars` (Nombre de voitures pouvant être garées dans le garage) montrent également une forte corrélation positive, indiquant que les maisons plus grandes et celles avec plus de places de garage tendent à avoir un prix de vente plus élevé.
- D'autres variables telles que `GarageArea`, `TotalBsmtSF` (Surface totale du sous-sol), et `1stFlrSF` (Surface du premier étage) montrent aussi des corrélations positives significatives avec le prix de vente.
- On peut observer quelques variables surprenantes : il semblerait qu'augmenter la taille du sous-sol de type 2 (`BsmtFinSF2`) diminue le prix, bien qu'une manière minime. La taille de la cuisine (`KitchenAbvGr`) semble également avoir cet effet.

Ce graphique aide à comprendre quelles caractéristiques des maisons ont le plus d'impact sur leur valeur marchande. Il oriente les décisions dans l'analyse des données immobilières et l'estimation des prix de vente. La qualité générale de la maison et sa superficie habitable sont les facteurs les plus importants, suivis de près par la taille du garage.

```python
numerical_features = ['SalePrice', 'LotFrontage', 'LotArea', 'GrLivArea', 'GarageArea']

# Créer le pairplot
g = sns.pairplot(train[numerical_features], corner=True)

# Ajouter des lignes de tendance
for i, j in zip(*np.tril_indices_from(g.axes, -1)):
    sns.regplot(x=numerical_features[j], y=numerical_features[i], data=train,
                ax=g.axes[i, j], scatter=False, line_kws={"color":"red"})

plt.suptitle('Visualisation des liens entre plusieurs variables numériques avec lignes de tendance', y=1.02)
plt.show()
```

Ces diagrammes montrent la relation entre plusieurs variables numériques du dataset. La première colonne, qui lie le prix de vente aux autres variables, est ici la plus intéressante. La ligne rouge indique la tendance des prix en fonction des valeurs des variables. On peut observer une forte linéarité entre ces paires de variables, car les lignes résultantes et leur degré de précision, indiqué par la zone en rose, sont très serrés.

Cela signifie qu'un modèle de régression linéaire peut fonctionner correctement sur ces données. En revanche, des modèles plus avancés comme XGBoost capables de capturer des relations non linéaires, n'ont pas besoin de cette hypothèse de linéarité. Cependant, pour que le modèle de base prévu (régression linéaire) soit significatif, il est crucial que cette hypothèse de linéarité soit validée.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(train[numerical_features].corr(), annot=True, cmap='coolwarm', center=0, mask=np.triu(train[numerical_features].corr()))
plt.title('Correlation matrix of numerical features', fontsize=14)
plt.show()
```

Cette heatmap représente la matrice de corrélation des principales caractéristiques numériques du dataset, y compris le prix de vente des maisons (SalePrice). La corrélation entre les variables est indiquée par des valeurs comprises entre -1 et 1, où des valeurs proches de 1 indiquent une forte corrélation positive et des valeurs proches de -1 indiquent une forte corrélation négative.

De manière générale, la heatmap montre que certaines caractéristiques ont des relations significatives avec le prix de vente. Par exemple, la surface habitable au-dessus du sol (GrLivArea) présente une corrélation positive élevée avec le prix de vente, avec une valeur de 0.71. Cela suggère que plus la surface habitable est grande, plus le prix de vente tend à être élevé. De même, la surface du garage (GarageArea) montre une corrélation notable avec le prix de vente, avec une valeur de 0.62, indiquant que les maisons avec des garages plus grands ont tendance à se vendre à des prix plus élevés.

Par ailleurs, il existe des corrélations modérées entre certaines autres caractéristiques, telles que la façade du terrain (LotFrontage) et la superficie du terrain (LotArea). Ces relations suggèrent que des terrains plus grands tendent à avoir des façades plus larges. De plus, la corrélation entre la surface habitable et la surface du garage montre que les maisons avec plus d'espace habitable ont également tendance à avoir des garages plus grands.

En résumé, cette heatmap permet de visualiser rapidement les relations linéaires entre les différentes caractéristiques du dataset. Les fortes corrélations entre certaines caractéristiques et le prix de vente indiquent des variables clés qui influencent le prix des maisons et qui peuvent être particulièrement utiles pour la modélisation prédictive.

# Création du modèle baseline : phase d’idéation

Le modèle baseline est une référence simple et initiale utilisée pour comparer les performances des modèles plus sophistiqués que l'on développera par la suite. Ce modèle de base n'a pas besoin d'être complexe ; il peut s'agir d'un modèle trivial comme un modèle de régression linéaire simple.

L'objectif principal de la création d'un modèle baseline est de fournir un point de départ pour évaluer l'efficacité des modèles plus avancés. En ayant une référence de base, il devient possible de déterminer si les nouveaux modèles apportent une amélioration significative ou non. Sans un modèle baseline, il serait difficile de juger la performance réelle des modèles plus complexes.

Un modèle baseline est utile pour plusieurs raisons. Premièrement, il permet de détecter rapidement des problèmes dans les données ou dans le processus de modélisation. Si un modèle avancé ne parvient pas à surpasser le modèle baseline, cela peut indiquer des problèmes tels que des données mal préparées ou des modèles mal configurés. Deuxièmement, il aide à définir des attentes réalistes pour les performances des modèles plus avancés. Enfin, il fournit une méthode simple pour valider les premières étapes du projet et s'assurer que les efforts déployés vont dans la bonne direction. Nous souhaitons aussi exploiter ce modèle pour offrir un service en ligne afin de fournir une estimation rapide et facile pour l'utilisateur.

## Création de la colonne utilisable pour le résultat voulu.

Afin de mesurer les performances de nos modèles, il est nécessaire de définir une métrique. Dans notre cas, nous avons choisi d'utiliser le Root Mean Squared Error (RMSE) entre le logarithm de prix de vente réel provenant du dataset et le logarithme du prix de vente estimé par notre modèle.

Les raisons sont les suivantes :

- **Adaptabilité du RMSE** : Le RMSE est une mesure particulièrement adaptée pour les tâches de régression car il donne une indication claire de l'erreur moyenne en termes d'unités dans le jeu de données. Cette métrique est sensible aux erreurs importantes car elle élève les erreurs au carré avant de prendre la racine carrée, ce qui la rend très utile pour identifier les modèles qui pourraient sous-performer de manière significative sur certains échantillons du jeu de données. Le RMSE peut donc aider à optimiser notre modèle en mettant l'accent sur la minimisation des grandes erreurs, ce qui est important dans des applications pratiques comme l'estimation des prix immobiliers où ces grandes erreurs peuvent être très coûteuses.

- **Utilité de la prise du logarithme** : L'utilisation du logarithme des prix de vente dans le calcul du RMSE offre plusieurs avantages significatifs. D'abord, elle aide à normaliser la distribution des prix, souvent asymétrique et étalée, en une distribution plus symétrique et serrée, réduisant ainsi l'influence des valeurs extrêmes ou des outliers. Cela rend l'évaluation plus robuste et moins sensible aux variations extrêmes de prix, qui pourraient autrement dominer l'erreur moyenne calculée. De plus, en comparant les logarithmes, nous mesurons les erreurs en termes de pourcentages plutôt qu'en valeurs absolues, ce qui est plus pertinent lorsque les gammes de prix varient largement entre les observations, comme c'est souvent le cas dans l'immobilier. Ainsi, le RMSE sur le log des prix fournit une idée plus précise de l'erreur relative, permettant une évaluation plus équitable entre des biens de différentes valeurs.



```python
train['LogSalePrice'] = np.log(train['SalePrice'])

# si nécessaire, vous pouvez obtenir le prix de vente réel a partir du log de cette manière : 
# Ce code permet de reverser le logarithme, et est donc utilisé lors de l'envoi des résultats sur Kaggle. 
# train['SalePrice'] = np.exp(train['LogSalePrice'])
```

```python
# Current main baseline model

numerical_cols = ['YearBuilt', 'LotArea', 'GrLivArea', 'TotRmsAbvGrd']
categorical_cols = ['BldgType', 'Neighborhood']

# Step 2: Preprocessing
# For numerical features: scale
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# For categorical features: apply one-hot encoding
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
    ('scaler', StandardScaler(with_mean=False))
])

# Combine preprocessors in a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Step 3: Create the regression pipeline
model_linear = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Step 4: Splitting the data into training and testing sets
all_cols = numerical_cols + categorical_cols
X = train[all_cols]
y = train['LogSalePrice']
# Cross-validation
scores = cross_val_score(model_linear, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)

# Print the results
print("Cross-validated RMSE scores:", rmse_scores)
print("Mean RMSE:", np.mean(rmse_scores))
```

#### Feature engineering :

Le feature engineering est un processus essentiel en data science, consistant à créer de nouvelles colonnes (ou features) à partir des colonnes existantes d'un dataset. Cette étape permet d'extraire des informations pertinentes et de capturer des relations complexes, améliorant ainsi la précision des modèles de machine learning. En générant de nouvelles features, on enrichit les données et on offre de meilleures perspectives aux modèles, augmentant leur performance, réduisant les erreurs et apportant des insights plus profonds et précis.

Nous avons introduit deux nouvelles features : « `AgeAtSale` » et « `YearsSinceRemodel` ».

- `AgeAtSale` (Âge de la maison au moment de la vente) : Calculée en soustrayant l'année de construction de l'année de vente, cette feature représente l'âge de la maison au moment de sa vente. L'âge d'une maison influence sa valeur, les maisons plus anciennes pouvant nécessiter des réparations, ce qui peut diminuer leur prix, ou avoir une valeur historique ajoutée.
- `YearsSinceRemodel` (Nombre d'années depuis la dernière rénovation) : Obtenue en soustrayant l'année de la dernière rénovation de l'année de vente, cette feature mesure le temps écoulé depuis la dernière rénovation majeure. Les rénovations récentes augmentent généralement la valeur d'une maison en améliorant son état et ses équipements, permettant de mieux estimer le prix de vente basé sur les améliorations récentes.
En intégrant ces nouvelles features, nous améliorons la pertinence et l'efficacité de nos modèles de machine learning.

```python
def house_age_at_sale(df):
    df['AgeAtSale'] = df['YrSold'] - df['YearBuilt']
    return df

def time_since_remod(df):
    df['YearsSinceRemodel'] = df['YrSold'] - df['YearRemodAdd']
    return df
```

```python
feature_engineering_transformers = [
    ('house_age_at_sale', FunctionTransformer(house_age_at_sale)),
    ('time_since_remod', FunctionTransformer(time_since_remod)),
]

# Create a custom transformer to apply the feature engineering transformations
class FeatureEngineering(BaseEstimator, TransformerMixin):
    def __init__(self, transformers):
        self.transformers = transformers

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        for name, transformer in self.transformers:
            X = transformer.transform(X)
        return X

# Instantiate the custom transformer
feature_engineering_step = FeatureEngineering(feature_engineering_transformers)
```

### Pipeline

```python
# Defining the imputer for NaN values. 
# Doing this in multiple steps allows us to define different imputers if they need to be filled differently. 

fill_numerical_features = []
mean_numerical_feature = [col for col in numerical_cols if col not in fill_numerical_features]

# Transformateurs pour les données numériques
mean_numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="mean"))
])
fill_numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value=0))
])

fill_categorical_nominal_cols = ['MiscFeature', 'Alley', 'MasVnrType']
mean_categorical_nominal_cols = [col for col in categorical_nominal_cols if col not in fill_categorical_nominal_cols]

# Transformateurs pour les données catégorielles nominales
categorical_nominal_transformer_fill = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='NA')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

categorical_nominal_transformer_mean = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

fill_categorical_ordinal_cols = ['PoolQC', 'Fence']
mean_categorical_ordinal_cols = [col for col in categorical_ordinal_cols if col not in fill_categorical_ordinal_cols]

# Transformateurs pour les données catégorielles ordinales
categorical_ordinal_transformer_fill = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('encoder', OrdinalEncoder(categories=[ordinal_mappings[col] for col in fill_categorical_ordinal_cols]))
])

categorical_ordinal_transformer_mean = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(categories=[ordinal_mappings[col] for col in mean_categorical_ordinal_cols]))
])

# Création du transformateur de colonnes
preprocessor = ColumnTransformer([
    ('num_fill', fill_numeric_transformer, fill_numerical_features),
    ('num_mean', mean_numeric_transformer, mean_numerical_feature),
    ('cat_nominal_fill', categorical_nominal_transformer_fill, fill_categorical_nominal_cols),
    ('cat_nominal_mean', categorical_nominal_transformer_mean, mean_categorical_nominal_cols),
    ('cat_ordinal_fill', categorical_ordinal_transformer_fill, fill_categorical_ordinal_cols),
    ('cat_ordinal_mean', categorical_ordinal_transformer_mean, mean_categorical_ordinal_cols)
], remainder='drop')

# Assemblage de la pipeline de prétraitement
preprocessing_pipeline = Pipeline([
    ('feature_engineering', feature_engineering_step),
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler())
])

# Ajustement et transformation du DataFrame
X_train = preprocessing_pipeline.fit_transform(train)

# Vérification de la pipeline
preprocessing_pipeline
```

## Pipeline

Notre pipeline comprend plusieurs étapes clés :
1. Feature Engineering : Nous commençons par créer des features supplémentaires qui peuvent fournir des informations supplémentaires pour le modèle.
2. Traitement des Données Manquantes : Ensuite, nous définissons comment traiter les données manquantes selon le type de données (numérique, ordinale ou catégorique) :
   1. Valeurs numériques : Les valeurs manquantes sont remplacées par 0. Par exemple, si "GarageCar" est nul, cela signifie que la maison n'a pas de garage, donc la valeur est 0.
   2. Valeurs catégoriques : Les valeurs manquantes sont remplacées par 'NA' pour les non-ordinales et 'Missing' pour les ordinales. Cela indique que la maison ne possède pas les éléments nécessaires pour entrer dans une des catégories définies.
3. Encodage des Catégories : Une fois les valeurs manquantes traitées, nous appliquons un OneHotEncoder aux colonnes catégoriques non ordinales et un OrdinalEncoder aux colonnes ordinales, préservant ainsi l'ordre des catégories.
4. Normalisation des Données : Enfin, nous utilisons un StandardScaler pour normaliser les données. Cela recentre les valeurs de chaque colonne autour de 0, améliorant la performance de la plupart des modèles de machine learning.

Ce pipeline garantit une préparation optimale des données, augmentant ainsi la précision et l'efficacité des modèles.

## Entrainer un modèle XGBoost

XGBoost, ou Extreme Gradient Boosting, est une bibliothèque de machine learning réputée pour ses performances élevées et sa rapidité d'exécution lors de l'entraînement des arbres de décision par gradient boosting. Conçu pour optimiser les ressources computationnelles, XGBoost offre des modèles précis et une exécution rapide, ce qui le rend idéal pour les tâches sur données structurées. XGBRegressor, une implémentation spécifique pour les problèmes de régression, bénéficie de fonctionnalités avancées telles que la régularisation, la construction parallèle d'arbres, et une gestion efficace de grands ensembles de données, le rendant très polyvalent et efficace.

Le Boosting est une technique d'ensemble utilisée en machine learning pour améliorer les performances prédictives d'un modèle. Il consiste à entraîner plusieurs modèles faibles de manière séquentielle, où chaque nouveau modèle corrige les erreurs de son prédécesseur. Cette approche permet de combiner plusieurs modèles pour obtenir une meilleure précision globale. XGBoost utilise ce concept en implémentant un gradient boosting avancé, optimisant ainsi le processus d'apprentissage grâce à des améliorations significatives. Parmi ces améliorations, on trouve la régularisation, qui aide à prévenir le surapprentissage, et la construction parallèle des arbres, qui accélère le processus d'entraînement. De plus, XGBoost est capable de gérer efficacement des ensembles de données de grande taille, le rendant particulièrement adapté pour les tâches de régression et de classification sur des données volumineuses.

Les hyperparamètres suivants sont utilisés pour ajuster le modèle XGBoostRegressor afin d'optimiser ses performances. Voici une description de chaque paramètre :

- **max_depth** : La profondeur maximale d'un arbre. Une valeur plus élevée permet au modèle de capturer des relations plus complexes, mais peut aussi entraîner un surapprentissage. Les valeurs sont généralement comprises entre 3 et 10.
- **learning_rate** (ou *eta*) : Le taux d'apprentissage réduit le poids de chaque arbre nouvellement ajouté. Une valeur plus faible nécessite plus d'arbres mais peut améliorer la précision. Le taux d'apprentissage varie généralement entre 0,01 et 0,3.
- **n_estimators** : Le nombre d'arbres à construire. Un nombre plus élevé d'estimateurs améliore la performance du modèle, mais augmente également le temps de calcul. Ici, le nombre est compris entre 5 et 500.
- **subsample** : La proportion d'échantillons de formation à utiliser pour chaque arbre. La sous-échantillonnage (subsampling) aide à prévenir le surapprentissage. Les valeurs vont de 0,5 à 1,0.
- **colsample_bytree** : La proportion de colonnes (caractéristiques) à utiliser pour chaque arbre. La sous-échantillonnage des colonnes permet également de réduire le surapprentissage. Les valeurs varient entre 0,5 et 1,0.
- **alpha** : Le terme de régularisation L1 (lasso). Augmenter cette valeur rend le modèle plus conservateur en ajoutant une pénalité pour les poids élevés. Les valeurs sont comprises entre 0 et 1.
- **lambda** : Le terme de régularisation L2 (ridge). Comme pour alpha, augmenter cette valeur rend le modèle plus conservateur. Les valeurs vont de 0 à 1.
- **min_child_weight** : La somme minimale des poids (hessians) nécessaire dans un nœud enfant. Une valeur plus élevée rend le modèle plus conservateur en empêchant la création de nœuds avec peu d'instances. Les valeurs varient entre 1 et 10.
- **gamma** : La réduction minimale de la perte requise pour effectuer une partition supplémentaire sur un nœud feuille. Des valeurs plus élevées rendent l'algorithme plus conservateur en nécessitant une plus grande réduction de la perte avant de créer une nouvelle division. Les valeurs varient entre 0 et 0,5.


```python
import optuna
import xgboost as xgb
import warnings
warnings.filterwarnings("ignore")

X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Define RMSE scorer
rmse_scorer = make_scorer(mean_squared_error, squared=False)

def objective(trial):
    param = {
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.3),
        'n_estimators': trial.suggest_int('n_estimators', 5, 500), 
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'alpha': trial.suggest_float('alpha', 0, 1),
        'lambda': trial.suggest_float('lambda', 0, 1),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0, 0.5)
    }
    
    # Create and train the model
    model = xgb.XGBRegressor(
        objective='reg:squarederror',
        booster='gbtree',
        verbosity=0,
        random_state=42,
        **param
    )
    
    # Use cross_val_score for cross-validation
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Return the mean RMSE
    return np.mean(scores)

# Create a study and optimize the objective function
study_xgbregressor = optuna.create_study(direction='minimize')
study_xgbregressor.optimize(objective, n_trials=100) # Similarly, you'd want to increase this number to like, 100-250

# Print best parameters
print('Best trial:')
trial = study_xgbregressor.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```

## Entrainer un modèle RandomForestRegressor

L'algorithme RandomForestRegressor de la bibliothèque scikit-learn est un modèle de machine learning basé sur les arbres de décision. Il utilise une approche d'ensemble pour améliorer la précision des prédictions. En créant une multitude d'arbres de décision indépendants sur des sous-échantillons du jeu de données, il moyenne les prédictions de ces arbres pour fournir une estimation finale plus robuste et moins sujette à l'overfitting. Cette technique permet de capturer les relations non linéaires dans les données et est particulièrement efficace pour les problèmes de régression avec de nombreux attributs.

Les hyperparamètres suivants sont utilisés pour ajuster le modèle RandomForestRegressor afin d'optimiser ses performances. Voici une description de chaque paramètre :
- bootstrap : Ce paramètre détermine si les échantillons bootstrap sont utilisés lors de la construction des arbres. Si bootstrap est défini sur True, un échantillonnage avec remise est effectué pour construire chaque arbre. Si False, l'ensemble de données complet est utilisé pour chaque arbre.
- n_estimators : Ce paramètre spécifie le nombre d'arbres dans la forêt. Plus le nombre d'arbres est élevé, plus les performances du modèle peuvent s'améliorer, mais cela augmente également le temps de calcul.
- max_depth : La profondeur maximale des arbres de décision. Limiter la profondeur des arbres peut aider à réduire le surapprentissage (overfitting). Si None, les nœuds sont étendus jusqu'à ce que toutes les feuilles soient pures ou contiennent moins de min_samples_split échantillons.
- min_samples_split : Le nombre minimum d'échantillons requis pour diviser un nœud interne. Un nombre plus élevé empêche les arbres de devenir trop complexes et aide à éviter le surapprentissage.
- min_samples_leaf : Le nombre minimum d'échantillons qu'une feuille peut contenir. Ce paramètre aide également à lisser le modèle, surtout dans les grandes données déséquilibrées.
- max_features : Le nombre de caractéristiques à considérer pour trouver la meilleure division à chaque nœud. Il peut être un entier, un flottant, ou prendre les valeurs 'auto', 'sqrt' (la racine carrée du nombre total de caractéristiques), ou 'log2' (le logarithme en base 2 du nombre total de caractéristiques). Limiter le nombre de caractéristiques examinées à chaque division peut réduire la variance du modèle.
- max_samples : Si bootstrap est défini sur True, ce paramètre spécifie le nombre maximum d'échantillons à tirer pour former chaque arbre. Il peut être défini comme un entier ou une fraction du nombre total d'échantillons. Cela permet  de contrôler la taille des sous-échantillons utilisés pour entraîner chaque arbre.


```python

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Définition du scorer RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# Fonction objective pour Optuna
def objective(trial):
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 5, 10),
        'max_depth': trial.suggest_int('max_depth', 3, 30),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        
        'bootstrap': bootstrap,
    }
    
    # Ajouter max_samples seulement si bootstrap est True
    if bootstrap:
        param['max_samples'] = trial.suggest_float('max_samples', 0.5, 1.0)
    
    # Création et entraînement du modèle
    model = RandomForestRegressor(
        random_state=42,
        **param
    )
    
    # Utilisation de cross_val_score pour la validation croisée
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Retourne la moyenne du RMSE
    return np.mean(scores)

# Création d'une étude et optimisation de la fonction objective
study_randomForest = optuna.create_study(direction='minimize')
study_randomForest.optimize(objective, n_trials=100)  

# Affichage des meilleurs paramètres
print('Best trial:')
trial = study_randomForest.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```

## Entrainer un modèle  Support Vector Regression  (SVR)

Le modèle Support Vector Regression (SVR) de scikit-learn est une extension de la méthode Support Vector Machine (SVM) utilisée pour les problèmes de régression. Contrairement à la classification où l'objectif est de séparer les classes avec une marge maximale, le SVR cherche à trouver une fonction qui dévie le moins possible des cibles réelles par une marge ε. Les points de données en dehors de cette marge sont pénalisés, ce qui permet de créer un modèle robuste et capable de généraliser sur des données non vues.

Les hyperparamètres suivants sont utilisés pour ajuster le modèle SVR afin d'optimiser ses performances. Voici une description de chaque paramètre :

- C : Ce paramètre de régularisation contrôle le compromis entre un modèle simple et les erreurs de régulation sur les points de données d'entraînement. Une valeur de C élevée permet au modèle de bien s'ajuster aux points de données d'entraînement, mais peut conduire à un surapprentissage. Une valeur de C plus faible conduit à un modèle plus régulier et moins susceptible de surapprendre.
- epsilon : Ce paramètre définit la largeur du tube ε autour de la fonction de régression où les erreurs ne sont pas pénalisées. Cela permet de contrôler la sensibilité du modèle aux erreurs sur les données d'entraînement. Un epsilon plus grand rend le modèle plus tolérant aux petites erreurs, tandis qu'un epsilon plus petit permet de réduire les erreurs d'entraînement, potentiellement au détriment de la généralisation.
- kernel : Ce paramètre spécifie le type de noyau à utiliser dans l'algorithme. Le choix du noyau a un impact significatif sur la capacité du modèle à capturer les relations complexes dans les données. Les options incluent :
  -'linear' : Utilise un noyau linéaire.
  - 'poly' : Utilise un noyau polynomial.
  - 'rbf' : Utilise un noyau de fonction de base radiale (Gaussian).
  - 'sigmoid' : Utilise un noyau sigmoïde.
- gamma : Ce paramètre est le coefficient du noyau pour les noyaux 'rbf', 'poly' et 'sigmoid'. Il peut être défini comme 'scale' (par défaut, 1 / (n_features * X.var())) ou 'auto' (1 / n_features), ou une valeur flottante. Il contrôle l'influence d'un seul point de donnée. Des valeurs de gamma élevées peuvent mener à un modèle plus flexible mais sujet au surapprentissage.
- degree : Ce paramètre est utilisé uniquement avec le noyau polynomial et détermine le degré du polynôme. Il doit être un entier non négatif et est ignoré par les autres noyaux.


```python
import optuna
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.svm import SVR
import numpy as np

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Définition du scorer RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# Fonction objective pour Optuna
def objective(trial):
    param = {
        'C': trial.suggest_loguniform('C', 1e-3, 1e3),
        'epsilon': trial.suggest_loguniform('epsilon', 1e-3, 1e1),
        'kernel': trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid']),
    }
    
    # Si le kernel est poly ou rbf, nous devons ajouter le paramètre gamma
    if param['kernel'] in ['poly', 'rbf', 'sigmoid']:
        param['gamma'] = trial.suggest_categorical('gamma', ['scale', 'auto'])
    
    # Si le kernel est poly, nous devons également ajouter le paramètre degree
    if param['kernel'] == 'poly':
        param['degree'] = trial.suggest_int('degree', 2, 5)
    
    # Création et entraînement du modèle
    model = SVR(**param)
    
    # Utilisation de cross_val_score pour la validation croisée
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Retourne la moyenne du RMSE
    return np.mean(scores)

# Création d'une étude et optimisation de la fonction objective
study_svr = optuna.create_study(direction='minimize')
study_svr.optimize(objective, n_trials=100)

# Affichage des meilleurs paramètres
print('Best trial:')
trial = study_svr.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```

## Entrainnement du modèle Elastic Net Regression

Le modèle Elastic Net Regression de scikit-learn est une méthode de régression linéaire qui combine les propriétés de la régularisation L1 (Lasso) et L2 (Ridge). Il vise à équilibrer les avantages des deux types de régularisation pour obtenir un modèle plus robuste et flexible. Cette combinaison permet de gérer les données ayant des caractéristiques corrélées et de sélectionner automatiquement les caractéristiques les plus pertinentes en les réduisant à zéro si nécessaire.

Les hyperparamètres suivants sont utilisés pour ajuster le modèle ElasticNet afin d'optimiser ses performances. Voici une description de chaque paramètre :

- alpha : Ce paramètre contrôle l'intensité de la régularisation. Une valeur plus élevée d'alpha augmente la régularisation, ce qui conduit à des coefficients plus petits et à un modèle plus simple. Cela peut aider à éviter le surapprentissage (overfitting), mais une régularisation excessive peut également entraîner un sous-apprentissage (underfitting). La régularisation aide à rendre le modèle plus général en réduisant la variance (machinelearningcompass) (machinelearningcompass).
- l1_ratio : Ce paramètre définit le mélange des régularisations L1 et L2. Une valeur de l1_ratio de 0 correspond à une régularisation de type Ridge (L2), tandis qu'une valeur de 1 correspond à une régularisation de type Lasso (L1). Les valeurs intermédiaires offrent un équilibre entre les deux. Un l1_ratio plus proche de 1 favorise la régularisation L1, conduisant à des modèles plus parcimonieux (avec des coefficients réduits à zéro), tandis qu'un l1_ratio plus proche de 0 favorise la régularisation L2, qui tend à distribuer la pénalité sur tous les coefficients (machinelearningcompass) (machinelearningcompass).
L'Elastic Net est particulièrement utile pour les jeux de données avec de nombreuses variables prédictives, surtout lorsque ces variables sont corrélées. En ajustant ces paramètres, vous pouvez contrôler la complexité du modèle et sa capacité à généraliser à de nouvelles données, obtenant ainsi des prédictions plus fiables et précises.


```python
import optuna
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.linear_model import ElasticNet
import numpy as np

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed_df, train['LogSalePrice'], test_size=0.2, random_state=42)

# Définition du scorer RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)

# Fonction objective pour Optuna
def objective(trial):
    param = {
        'alpha': trial.suggest_loguniform('alpha', 1e-5, 1e1),
        'l1_ratio': trial.suggest_float('l1_ratio', 0, 1)
    }
    
    # Création et entraînement du modèle
    model = ElasticNet(random_state=42, **param)
    
    # Utilisation de cross_val_score pour la validation croisée
    scores = cross_val_score(model, X_train, y_train, scoring=rmse_scorer, cv=5)
    
    # Retourne la moyenne du RMSE
    return np.mean(scores)

# Création d'une étude et optimisation de la fonction objective
study_elastic = optuna.create_study(direction='minimize')
study_elastic.optimize(objective, n_trials=100)

# Affichage des meilleurs paramètres
print('Best trial:')
trial = study_elastic.best_trial

print('  Value: {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))
```
